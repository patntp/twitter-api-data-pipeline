```python
import tweepy
import pandas as pd
import psycopg2 as ps
```

```python
CONSUMER_KEY="TWITTER_API_KEY"
CONSUMER_SECRET="TWITTER_API_KEY_SECRET"
ACCESS_TOKEN="ACCESS_TOKEN"
ACCESS_TOKEN_SECRET="ACCESS_TOKEN_SECRET"
```

```python
# Authenticate to Twitter
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN,ACCESS_TOKEN_SECRET)
api = tweepy.API(auth)
```

```python
# test authentication
try:
    api.verify_credentials()
    print("Authentication OK")
except:
    print("Error during authentication")
```

```python
df=pd.DataFrame(columns=['Name','Tweet Volume','Created at'])
df
```

```python
t=api.get_place_trends(id=23424960) #Thailand WOEID
for trend in t[0]['trends']:
    df=df.append({'Name':trend['name'],'Tweet Volume':trend['tweet_volume'],'Created at':t[0]['created_at']},ignore_index=True)
```

```python
df
```

```python
#Drop rows that does not have tweet volume
df.dropna(inplace=True)
df
```

```python
#Sort dataframe by Tweet Volume descendingly
df.sort_values(by='Tweet Volume',ascending=False,inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
def connect_to_db(host_name, dbname, port, username, password):
    try:
        conn = ps.connect(host=host_name, database=dbname, user=username, password=password, port=port)
    except ps.OperationalError as e:
        raise e
    else:
        print('Connected!')
    return conn
```

```python
def create_table(curr):
    create_table_command=("""CREATE TABLE IF NOT EXISTS trends(name VARCHAR(300) PRIMARY KEY,tweet_volume BIGINT NOT NULL,created_at DATE NOT NULL)""")
    curr.execute(create_table_command)
```

```python
def update_db(curr,df):
    tmp_df=pd.DataFrame(columns=['Name','Tweet Volume','Created at'])
    
    for i,row in df.iterrows():
        if check_if_name_exists(curr,row['Name']):
            update_row(curr,row['Name'],row['Tweet Volume'],row['Created at'])
        else:
            tmp_df=tmp_df.append(row)
            
    return tmp_df
```

```python
def check_if_name_exists(curr,name):
    query=("""SELECT name FROM trends WHERE name=%s""")
    curr.execute(query,(name,))
    return curr.fetchone() is not None    #return nothing if name has not existed
```

```python
def update_row(curr,name,tweet_volume,created_at):
    query=("""UPDATE trends SET tweet_volume=%s, created_at=%s WHERE name=%s;""")
    vars_to_update=(tweet_volume,created_at,name)
    print(vars_to_update) #show names that will be updated (already existed)
    curr.execute(query,vars_to_update)
```

```python
def append_from_df_to_db(curr,df):
    for i,row in df.iterrows():
        insert_into_table(curr,row['Name'],row['Tweet Volume'],row['Created at'])
```

```python
def insert_into_table(curr,name,tweet_volume,created_at):
    query=("""INSERT INTO trends (name,tweet_volume,created_at) VALUES(%s,%s,%s);""")
    row_to_insert=(name,tweet_volume,created_at)
    curr.execute(query,row_to_insert)
```
